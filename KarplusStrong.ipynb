```python
import numpy as np
```

```python
#Benchmark frequency value for adjusting the hold factor
REF_FREQ_HZ = 440
```

```python
def generate_karplus_strong(frequency = 440, sample_rate = 44100, hold_factor = 0.99, length_sec=2):
    
    input_waveform_length = np.floor(sample_rate/frequency)
    
    #Excitation waveform
    input_waveform = (2*np.random.rand(int(input_waveform_length)))-1
    
    #Output signal initialization
    out_signal_length = int(length_sec*sample_rate)
    out_signal = np.zeros(out_signal_length)
    
    #Adjusting hold factor to prevent the effect of frequency on output signal length
    scaled_hold_factor = hold_factor**(REF_FREQ_HZ/frequency)
    
    #Generating output signal
    curr_index = 0
    n = 0 #iterations
    while(curr_index<out_signal_length):
        if(curr_index+input_waveform_length<out_signal_length):
            out_signal[int(curr_index):int(curr_index+input_waveform_length)] = input_waveform * (scaled_hold_factor**n)
            curr_index += input_waveform_length
            n += 1
        else:
            out_signal[int(curr_index):] = (input_waveform[:int(out_signal_length-curr_index)])* (scaled_hold_factor**n)
            break
    
    return out_signal
```

```python
synth_sound = generate_karplus_strong(100,44100,0.99,2)
```
